```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
dataframe= pd.read_csv('/content/drive/MyDrive/train.csv')
```

```python
dataframe.head()
```

B) Removing some of the units.

```python
dataframe['Mileage'] = dataframe['Mileage'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
dataframe['Engine'] = dataframe['Engine'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
dataframe['Power'] = dataframe['Power'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
dataframe['New_Price'] = dataframe['New_Price'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
dataframe.head()
```

```python
dataframe.isnull().sum()
```

A) Replacing missing values with median, mean and mode

```python
dataframe['Mileage'].fillna(dataframe['Mileage'].median(),inplace=True)
dataframe['Engine'].fillna(dataframe['Engine'].median(),inplace=True)
dataframe['Power'].fillna(dataframe['Power'].mean(),inplace=True)
dataframe['Seats'].fillna(dataframe['Seats'].mode()[0],inplace=True)

dataframe.drop('New_Price', axis=1, inplace=True)
```

```python
dataframe.isnull().sum()
```

```python
dataframe.head()
```

C) Changing the categorical variables into numerical one hot encoded value

```python
dataframe['Fuel_Type'].replace({'Diesel': 0, 'Petrol': 1, 'Electric': 2}, inplace=True)
dataframe['Transmission'].replace({'Manual': 0, 'Automatic': 1}, inplace=True)
```

```python
dataframe.head()
```

D) Creating one more feature

```python
from datetime import datetime
present = datetime.now().year
dataframe['Car_Age'] = present - dataframe['Year']
dataframe.head()
```

E) Set of operations
1) SELECT

```python
selected_columns = dataframe[['Mileage', 'Engine']]
selected_columns.head()
```

2) FILTER

```python
filter_data = dataframe[dataframe['Year'] > 2013]
filter_data.head()
```

3) RENAME

```python
dataframe.rename(columns={'Location': 'Place_Manufactured', 'Power': 'Horsepower'}, inplace=True)
dataframe.head()
```

4) MUTATE

```python
dataframe['Mileage_per_CC'] = dataframe['Mileage'] / dataframe['Engine']
dataframe.head()
```

5) ARRANGE

```python
arranged_data = dataframe.sort_values(by='Price')
arranged_data.head()
```
